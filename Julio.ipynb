```python
#1 Que hace este programa?
import csv

with open('tripdata.csv') as archivoCSV: 
    lector = csv.reader(archivoCSV, delimiter=',')
    c = 0
    tip=0
    for registro in lector:
        if c > 1 and registro[13]!='':
            tipRegistro=float(registro[13])
            if tipRegistro>tip:
                tip=tipRegistro
        c += 1
    
print(tip)
#Busca e imprime la propina(tip) mas alta.
#No es el tip pero no se que es. Es el RatecodeID
```

```python
#2 Que hace este programa?
import csv

with open('tripdata.csv') as archivoCSV: 
    lector = csv.reader(archivoCSV, delimiter=',')
    c = 0
    cantidad=0
    for registro in lector:
        if c > 1 and registro[13]!='':
            tipRegistro=float(registro[13])
            if tipRegistro==100:
                cantidad+=1
        c += 1
    
print(cantidad)
#Busca y cuenta cuantos registros tienen el valor 99(100) en la posicion 13
```

```python
#Programa Libre! Cuanto gana un taxista por distancia(en millas) promedio. 
import csv

with open('tripdata.csv') as archivoCSV:
    lector = csv.reader(archivoCSV, delimiter=',')
    montosPorMilla = []
    c = 0
    for registro in lector:
        if c > 1:
            # Distancia del viaje en millas
            distancia = float(registro[4])
            # hay distancias en las cuales el valor es 0
            if distancia == 0:
                distancia = 1
            # Monto total del viaje dividido por la distancia por viaje
            montosPorMilla.append(float(registro[16]) / distancia)
        c += 1
        
promedio = sum(montosPorMilla) / len(montosPorMilla)
print("Promedio un taxista gana unos ${:.2f} por cada milla.".format(promedio))
```

```python
#propina mas alta menor de 100
import csv

with open('tripdata.csv') as archivoCSV: 
    lector = csv.reader(archivoCSV, delimiter=',')
    c = 0
    tip=0
    for registro in lector:
        if c > 1 and registro[13]!='':
            tipRegistro=float(registro[13])
            if tipRegistro < 100.0 and tipRegistro>tip:
                tip=tipRegistro
        c += 1
    
print(tip)
```
