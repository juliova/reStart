```python
with open('objetojson') as file:
    for linea in file:
      datosDelArchivo = json.loads(linea)
      #print(type(datosDelArchivo))
      try:
        print("Nombre: {}, Cedula: {}".format(datosDelArchivo["nombre"], datosDelArchivo["cedula"]))
      except KeyError:
        print("Este registro: {} le faltaba una Key".format(datosDelArchivo["nombre"]))
```

```python
#try
valores=[1,2,3,4,5,6,"a",7,8,"zzzzz",9,34,65,76,87,98,54]
suma=0
for val in valores:
    print(val)
    try:
        suma+=val
    except TypeError:
        print("El elemento '{}' no se puede sumar por ser string".format(val))

print("la suma es: " + str(suma))
#Obtener los string de los valores
strings = [x for x in valores if type(x)==str]
print(strings)
```
